```python
import pandas as pd
```

```python
data = pd.read_csv('./weatherHistory_clean.csv')
data.info()
```

```python
data=data.drop("Formatted Date",axis=1)
```

```python
y=data['Summary']
y
```

```python
X= data.drop("Summary",axis=1)
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, bins=[-0.5, 0.5, 1.5, 2.5], color='red', rwidth=0.5)
plt.xlabel('Summary')
plt.ylabel('Number of Samples')
plt.xticks([0, 1, 2])  
plt.show()
```

```python
X.corr()
```

```python
data.corr()["Summary"].abs().sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split 

X_train,X_test,y_train,y_test = train_test_split(X, y,test_size=0.20,random_state=0) 
print(X_train.shape )
print(X_test.shape) 
print(y_train.shape) 
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn import metrics 
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score
```

```python
for k in range(3, 12, 2):  # test k = 3, 5, 7, 9, 11
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, y, cv=5)
    print(f"k = {k}, accuracy = {scores.mean():.4f}")
```

```python
classifier = KNeighborsClassifier(n_neighbors=5)
```

```python
model1=classifier.fit(X_train, y_train)
```

```python
y_pred1 = model1.predict(X_test) 
print("The accuracy is "+str(metrics.accuracy_score(y_test,y_pred1)*100)+"%") 
print(confusion_matrix(y_test, y_pred1))
```

```python
target_names = ['class 0', 'class 1','class 2'] 
print(classification_report(y_test, y_pred1, target_names=target_names)) 
```

### logistic regression

```python
from sklearn.linear_model import LogisticRegression 
logistic_regression1 = LogisticRegression(solver="liblinear", random_state=0) 
```

```python
model2 = logistic_regression1.fit(X_train, y_train)
```

```python
y_pred2 = model2.predict(X_test) 
```

```python
from sklearn import metrics 
import matplotlib.pyplot as plt  
from sklearn.metrics import confusion_matrix, classification_report 

print("The accuracy is "+str(metrics.accuracy_score(y_test,y_pred2)*100)+"%") 
print(confusion_matrix(y_test, y_pred2))  
target_names = ['class 0', 'class 1','class 2'] 
print(classification_report(y_test, y_pred2, target_names=target_names))  
```
