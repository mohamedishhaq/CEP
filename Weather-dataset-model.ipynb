## Neural Network

### 📦 Importing Libraries

In this section, we import the necessary libraries for building and evaluating a classification model using PyTorch and scikit-learn

```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

### 🔄 Converting Data to NumPy Arrays

Before feeding the data into a PyTorch model, we ensure that the features and labels are converted into NumPy arrays

```python
import numpy as np

# Ensure labels are converted to NumPy arrays
X_train_np = np.array(X_train)
X_test_np = np.array(X_test)
y_train_np = np.array(y_train)
y_test_np = np.array(y_test)
```

### 🔁 Converting NumPy Arrays to PyTorch Tensors

To train the model using PyTorch, we need to convert the NumPy arrays into PyTorch tensors:


```python
X_train_tensor = torch.from_numpy(X_train_np).float()
y_train_tensor = torch.from_numpy(y_train_np).long()
X_test_tensor = torch.from_numpy(X_test_np).float()
y_test_tensor = torch.from_numpy(y_test_np).long()
```

### 🧠 Defining the Neural Network Model

We define a simple feedforward neural network using PyTorch's `nn.Module`:


```python

# Neural Network Model
class Net(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(64, 32)
        self.relu2 = nn.ReLU()
        self.output = nn.Linear(32, output_dim)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.output(x)  # No softmax here; CrossEntropyLoss includes it
        return x
```

### ⚙️ Model Initialization and Setup


```python

# Initialize model
model = Net(input_dim=9, output_dim=3)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

### 🔄 Training Loop with Loss Tracking


```python
losses = []  # Store losses for plotting

for epoch in range(3000):
    model.train()
    outputs = model(X_train_tensor)
    loss = criterion(outputs, y_train_tensor)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    losses.append(loss.item())  # Record loss
    if (epoch+1) % 10 == 0:
        print(f'Epoch [{epoch+1}/100], Loss: {loss.item():.4f}')
```

### ✅ Model Evaluation on Test Set


```python
# Evaluation on test set
model.eval()
with torch.no_grad():
    outputs = model(X_test_tensor)
    _, predicted = torch.max(outputs, 1)

    accuracy = (predicted == y_test_tensor).float().mean()
    print(f"\nTest Accuracy: {accuracy.item() * 100:.2f}%")

    # Classification report
    from sklearn.metrics import classification_report
    print("\nClassification Report:\n", classification_report(y_test_tensor.numpy(), predicted.numpy()))
```

### 📉 Plotting Training Loss Over Epochs


```python
# Plotting the loss curve
plt.figure(figsize=(8, 5))
plt.plot(range(1, 3001), losses, marker='o')
plt.title('Training Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.tight_layout()
plt.show()
```
